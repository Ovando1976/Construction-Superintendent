```python

from tkinter import *
from tkinter import ttk
from PIL import Image, ImageTk

class ConstructionApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Construction Management App")
        self.master.geometry("800x600")
        self.master.config(bg="#f2f2f2")
        
        # Add app icon
        self.master.iconbitmap("icon.ico")
        
        # Add app title
        title_frame = Frame(self.master, bg="#dbaa1a")
        title_frame.pack(fill=X)
        
        title_label = Label(title_frame, text="Construction Management App", font=("Helvetica", 24), fg="white", bg="#dbaa1a")
        title_label.pack(side=LEFT, padx=20, pady=10)
        
        # Add chat bot avatar
        image = Image.open("avatar.png")
        image = image.resize((100, 100), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(image)
        avatar_label = Label(self.master, image=photo, bg="#f2f2f2")
        avatar_label.image = photo
        avatar_label.place(x=20, y=70)
        
        # Add tabs
        tabs = ttk.Notebook(self.master)
        tabs.pack(fill=BOTH, expand=1, padx=20, pady=10)
        
        home_tab = ttk.Frame(tabs)
        budget_tab = ttk.Frame(tabs)
        schedule_tab = ttk.Frame(tabs)
        employees_tab = ttk.Frame(tabs)
        tools_tab = ttk.Frame(tabs)
        
        tabs.add(home_tab, text="Home")
        tabs.add(budget_tab, text="Budget")
        tabs.add(schedule_tab, text="Schedule")
        tabs.add(employees_tab, text="Employees")
        tabs.add(tools_tab, text="Tools")
        
        # Add content to home tab
        home_frame = Frame(home_tab, bg="#fff8e5")
        home_frame.pack(fill=BOTH, expand=1)
        
        home_label = Label(home_frame, text="Welcome to the Construction Management App!", font=("Helvetica", 18), fg="#3d3d3d", bg="#fff8e5")
        home_label.pack(pady=20)
        
        home_text = """This app is designed to help you manage your construction project from start to finish. 
        Use the tabs above to access different features like budget management, scheduling, and more."""
        home_text_label = Label(home_frame, text=home_text, font=("Helvetica", 14), fg="#3d3d3d", bg="#fff8e5", justify="left")
        home_text_label.pack(padx=50, pady=10)
        
                # Add content to budget tab
        budget_frame = Frame(budget_tab, bg="#e8f0fa")
        budget_frame.pack(fill=BOTH, expand=1)
        
        budget_label = Label(budget_frame, text="Budget Management", font=("Helvetica", 24, "bold"), fg="#3c3c3c", bg="#e8f0fa")
        budget_label.pack(pady=30)
        
        # Add budget image
        budget_image = Image.open("budget.png")
        budget_image = budget_image.resize((300, 200), Image.ANTIALIAS)
        budget_photo = ImageTk.PhotoImage(budget_image)
        budget_image_label = Label(budget_frame, image=budget_photo, bg="#e8f0fa")
        budget_image_label.image = budget_photo
        budget_image_label.pack(pady=20)
        
        # Add budget content
        budget_text = """Use this tab to manage your construction budget. You can add expenses, 
        track spending, and view budget reports. The app allows you to set a budget for each category 
        of your project and track expenses against that budget. You can also view reports that show 
        spending by category, spending over time, and more."""
        
        budget_text_label = Label(budget_frame, text=budget_text, font=("Helvetica", 14), fg="#3c3c3c", bg="#e8f0fa", justify="left")
        budget_text_label.pack(padx=100, pady=20)
        
        # Add expenses button
        expenses_button = Button(budget_frame, text="Add Expenses", font=("Helvetica", 14, "bold"), bg="#326ac7", fg="white", padx=10, pady=5, bd=0)
        expenses_button.pack(side=LEFT, padx=100, pady=50)
        
        # Add reports button
        reports_button = Button(budget_frame, text="View Reports", font=("Helvetica", 14, "bold"), bg="#326ac7", fg="white", padx=10, pady=5, bd=0)
        reports_button.pack(side=LEFT, padx=100, pady=50)
        
        # Add content to schedule tab
        schedule_frame = Frame(schedule_tab, bg="#f2f2f2")
        schedule_frame.pack(fill=BOTH, expand=1)

        schedule_title_label = Label(schedule_frame, text="Schedule Management", font=("Helvetica", 18), fg="#333", bg="#f2f2f2")
        schedule_title_label.pack(pady=20)

        schedule_info_label = Label(schedule_frame, text="Use this tab to manage your construction schedule. You can add tasks, assign them to employees, and track progress.", font=("Helvetica", 14), fg="#333", bg="#f2f2f2", justify="left")
        schedule_info_label.pack(padx=50, pady=10)

        # Add content to employee tab
        employee_frame = Frame(employee_tab, bg="#f2f2f2")
        employee_frame.pack(fill=BOTH, expand=1)

        employee_title_label = Label(employee_frame, text="Employee Management", font=("Helvetica", 18), fg="#333", bg="#f2f2f2")
        employee_title_label.pack(pady=20)

        employee_info_label = Label(employee_frame, text="Use this tab to manage your construction employees. You can add employees, assign them to tasks, and track their progress.", font=("Helvetica", 14), fg="#333", bg="#f2f2f2", justify="left")
        employee_info_label.pack(padx=50, pady=10)

        # Add content to tool tab
        tool_frame = Frame(tool_tab, bg="#f2f2f2")
        tool_frame.pack(fill=BOTH, expand=1)

        tool_title_label = Label(tool_frame, text="Tool Management", font=("Helvetica", 18), fg="#333", bg="#f2f2f2")
        tool_title_label.pack(pady=20)

        tool_info_label = Label(tool_frame, text="Use this tab to manage your construction tools. You can add tools, track their usage, and schedule maintenance.", font=("Helvetica", 14), fg="#333", bg="#f2f2f2", justify="left")
        tool_info_label.pack(padx=50, pady=10)

        # Add content to equipment tab
        equipment_frame = Frame(equipment_tab, bg="#f2f2f2")
        equipment_frame.pack(fill=BOTH, expand=1)

        equipment_title_label = Label(equipment_frame, text="Equipment Management", font=("Helvetica", 18), fg="#333", bg="#f2f2f2")
        equipment_title_label.pack(pady=20)

        equipment_info_label = Label(equipment_frame, text="Use this tab to manage your construction equipment. You can add equipment, track their usage, and schedule maintenance.", font=("Helvetica", 14), fg="#333", bg="#f2f2f2", justify="left")
        equipment_info_label.pack(padx=50, pady=10)

        # Add content to chatbot tab
        chatbot_frame = Frame(chatbot_tab, bg="#f2f2f2")
        chatbot_frame.pack(fill=BOTH, expand=1)

        chatbot_title_label = Label(chatbot_frame, text="Chatbot", font=("Helvetica", 18), fg="#333", bg="#f2f2f2")
        chatbot_title_label.pack(pady=20)

        chatbot_info_label = Label(chatbot_frame, text="Use this tab to ask general construction related questions or diagnose issues in plumbing, carpentry, electricity, and HVAC.", font=("Helvetica", 14), fg="#333", bg="#f2f2f2", justify="left")
        chatbot_info_label.pack(padx=50, pady=10)

# Add tabs to main notebook
        tabs
```
